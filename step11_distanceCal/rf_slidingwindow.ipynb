```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

## For Spike


```python
output = pd.read_csv("/content/rf_spike.csv")
```

```python

output['Start position'] = output["window_pos"].str.split("_").str[0].astype(int)
output['RF distance'] = output["normalized_RF"] *100
output
```

```python
fig, ax = plt.subplots(figsize=(24, 6))

fig = sns.lineplot(data=output, x="Start position", y="RF distance", ax=ax)
fig.set(xlabel='Start position', ylabel='Persentage of normalised Robinson and Foulds distance')
fig.set_xticks(range(0,800,10))
fig.set_yticks(range(85,100,1))
ax.tick_params(axis='x', rotation=60)

plt.show()
```

```python
ax.figure.savefig('fig_spike.png',transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(24, 6))

fig = sns.scatterplot(data=output, x="Start position", y="RF distance", ax=ax)
fig.set(xlabel='Start position', ylabel='Persentage of normalised Robinson and Foulds distance')
fig.set_xticks(range(0,800,10))
fig.set_yticks(range(85,100,1))
ax.tick_params(axis='x', rotation=60)

plt.show()
```

```python
output["RF distance"].min()
```

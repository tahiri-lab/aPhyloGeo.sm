## Get association matrix

```python
df = df[['Strain','V_Accession_full','Host','Host_Accession']]
df
```

```python
with open('Bat_matrix.txt', 'r') as f:
    bat_label = [line.strip().split(' ')[0] for line in f][1:]
```

```python
bat_label
```

```python
len(bat_label)
```

```python
with open('V_matrix.txt', 'r') as f:
    v_label = [line.strip().split(' ')[0] for line in f][1:]
len(v_label)
```

```python
v_label
```

```python
bat_id_lt = [line.split('_')[0].split('.')[0] for line in bat_label]
print(bat_id_lt)
```

```python
v_id_lt = [line.split('_')[0].split('.')[0] for line in v_label]
print(v_id_lt)
```

```python
df_m = pd.DataFrame(index=bat_id_lt, columns=v_id_lt)
df_m = df_m.fillna(int(0))
df_m
```

```python
df
```

```python
# Iterate over each row
for index, row in df.iterrows():
    bat = row['Host_Accession'].split('.')[0].strip()
    virus = row['V_Accession_full'].split('_')[0].split('.')[0].strip()
    df_m.loc[bat,virus]=int(1)
    print(bat,virus)
```

```python
df_m
```

```python
df_m.to_csv('link.csv',index=True)
```
